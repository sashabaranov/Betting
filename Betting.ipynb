# Simulation of https://kaue.me/bet/

```python
import random
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
configs = [
    {"prize": 1.25, "prob": 0.7600},
    {"prize": 1.50, "prob": 0.6333},
    {"prize": 4.00, "prob": 0.2375},
    
    {"prize": 9.50, "prob": 0.02},
    {"prize": 19.0, "prob": 0.05},
    {"prize": 47.5, "prob": 0.1},
]
```

```python
random.seed(1)
def simulate(prize, prob, bets=500, balance=100., bet=1.0):
    balances = [balance]
    for _ in range(bets):
        balance -= 1.0
        if random.random() < prob:
            balance += prize
        balances.append(balance)
    return balances
```

```python
plt.figure(figsize=(16, 6))

for cfg in configs:
    for _ in range(2):
        prize, prob = cfg["prize"], cfg["prob"]
        dataset = simulate(prize, prob, bets=1000)
        sns.lineplot(range(len(dataset)), dataset, label=f"${prize} with prob {prob*100}%")
```
